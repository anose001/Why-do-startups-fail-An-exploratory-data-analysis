# Compiling a single data set from the single data files and clean the data

```python
import pandas as pd
import numpy as np
import re
pathnames=['StartUpFailure_Crunchbase_2000_UTF8.txt','StartUpFailure_Crunchbase_2001_UTF8.txt',
          'StartUpFailure_Crunchbase_2002_UTF8.txt','StartUpFailure_Crunchbase_2003_UTF8.txt',
          'StartUpFailure_Crunchbase_2004_UTF8.txt','StartUpFailure_Crunchbase_2005_UTF8.txt',
          'StartUpFailure_Crunchbase_2006_UTF8.txt','StartUpFailure_Crunchbase_2007_UTF8.txt',
          'StartUpFailure_Crunchbase_2008_UTF8.txt','StartUpFailure_Crunchbase_2009_UTF8.txt',
          'StartUpFailure_Crunchbase_2010_UTF8.txt','StartUpFailure_Crunchbase_2011_UTF8.txt',
          'StartUpFailure_Crunchbase_2012_UTF8.txt','StartUpFailure_Crunchbase_2013_UTF8.txt',
          'StartUpFailure_Crunchbase_2014_UTF8.txt','StartUpFailure_Crunchbase_2015_UTF8.txt',
          'StartUpFailure_Crunchbase_2016_UTF8.txt','StartUpFailure_Crunchbase_2017_UTF8.txt',
          'StartUpFailure_Crunchbase_2018_UTF8.txt','StartUpFailure_Crunchbase_2019_UTF8.txt']
pathnames=['StartUpFailure_Crunchbase_2000_UTF8.txt','StartUpFailure_Crunchbase_2001_UTF8.txt']

def compile_and_clean_data_to_one_file(pathnames):
    '''pathnames: list of pathnames (.txt extension) containing line by line data from crunchbase for each year
       return: df with compiled data (saved to csv file)
    '''
    for path in pathnames:
        print path
        # list with dictionaries will be fed to df. Each dictionary will be one row in the dataframe.
        l=[]
        # number of lines in file that is read in
        fo = open(path)
        num_lines = sum(1 for line in fo)
        # counter for number of companies in a given text file
        check_list = [str(number)+'.' for number in xrange(1,1100)]
        fo.close()
        # open file
        with open(path) as myfile:
            # the first 50 lines are the columns
            columns = [next(myfile).split('\n')[0] for x in xrange(50)]
            # initiate dictionary to save the information of each company
            dictionary = {}
            # iterate over all the lines in a given file
            for x in xrange(num_lines-50):
                # save data in a line in a string variable
                string = next(myfile).split('\n')[0]
                print string
                if string in check_list: next_column, dictionary =0, {}
                elif next_column == 49: l.append(dictionary)
                elif 'Logo of' in string: continue
                else: 
                    if string.decode('UTF-8') =='—'.decode('UTF-8'): string=np.nan
                    dictionary[columns[next_column]]=string
                    next_column+=1
        # create a dataframe from the list of libraries and save it as a csv file
        df=pd.DataFrame(l)
        df.to_csv(path.split('.')[0]+'.csv')

    #pathnames=['StartUpFailure_Crunchbase_2000_UTF8.csv','StartUpFailure_Crunchbase_2001_UTF8.csv']
    pathnames = [string.replace('.txt','.csv') for string in pathnames]
    #initiate the final dataframe
    path = pathnames[0]
    df_all = pd.read_csv(path, encoding='utf-8').drop(['Unnamed: 0'], axis=1)
    df_all['Closed Date Year'] =  path.split('_')[-2]
    #append other dataframes to initiated dataframe
    for path in pathnames[1:]:
        df_dummy = pd.read_csv(path, encoding='utf-8').drop(['Unnamed: 0'], axis=1)
        df_dummy['Closed Date Year'] =  path.split('_')[-2]
        df_all = df_all.append(df_dummy, ignore_index=True)
        
    # This line drops 30 lines which do have foreign currencies (it is faster to drop them then go back at time and find out the proper conversion rates for the currencies)
    # in them using a regex and stores it in a new data frame df2.
    df2 = df.drop(df[df['Total Funding Amount'].str.contains("CHF|CN|DKK|ISK|R|SEK|SGD", na=False, regex=True)]['Total Funding Amount'].index).copy()
    #remove unwanted characters which are not numbers 
    bad_chars = r'[^1234567890]'
    # This line removes all the characters that are not numbers in the total funding amount column
    df2['Total Funding Amount']=df2['Total Funding Amount'].dropna().apply(lambda s: re.sub(bad_chars, '', s))             

    # save all data into one csv file
    df_all.to_csv('Failed_StartUps_2000_2019_test.csv', encoding='utf-8')
    return df_all

df = compile_and_clean_data_to_one_file(pathnames)
```
